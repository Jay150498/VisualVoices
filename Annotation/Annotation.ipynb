```python
from skimage import io
image_path = '/content/drive/MyDrive/asl_dataset_preprocessed/7/hand2_7_bot_seg_5_cropped.jpeg'  # Replace with your image path
img = io.imread(image_path)
#image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis('off')
plt.show()

# Define the annotation coordinates
x1, y1 = 70, 0  # Top-left corner coordinates
x2, y2 = 340, 400  # Bottom-right corner coordinates

# Draw a bounding box on the image
annotated_image = cv2.rectangle(img.copy(), (x1, y1), (x2, y2), (255, 255, 255), thickness=2)

# Display the annotated image
annotated_image_rgb = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)
plt.imshow(annotated_image)
plt.axis('off')
plt.show()

output_path = '/content/drive/MyDrive/annotated_image/7/7_17.jpg'  # Replace with your desired output path
cv2.imwrite(output_path, annotated_image)
```
